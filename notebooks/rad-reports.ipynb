```python
from pathlib import Path
import polars as pl
import json
```

## RadGraph V1 (https://physionet.org/content/radgraph/1.0.0/)

```python
radgraph_train_path = Path(
    "../../data/radgraph-extracting-clinical-entities-and-relations-from-radiology-reports-1.0.0/train.json"
)
assert radgraph_train_path.exists(), "Train data file does not exist."

radgraph_dev_path = Path(
    "../../data/radgraph-extracting-clinical-entities-and-relations-from-radiology-reports-1.0.0/dev.json"
)
assert radgraph_dev_path.exists(), "Dev data file does not exist."

radgraph_test_path = Path(
    "../../data/radgraph-extracting-clinical-entities-and-relations-from-radiology-reports-1.0.0/test.json"
)
assert radgraph_test_path.exists(), "Test data file does not exist."

rad_graph_train = json.load(open(radgraph_train_path))
rad_graph_dev = json.load(open(radgraph_dev_path))
rad_graph_test = json.load(open(radgraph_test_path))
```

## RadGraph XL (https://physionet.org/content/radgraph-xl/1.0.0/)

```python
mimic_data_path = Path(
    "../../data/radgraph-xl-a-large-scale-expert-annotated-dataset-for-entity-and-relation-extraction-from-radiology-reports-1.0.0/mimic-radgraph-XL.jsonl"
)
assert mimic_data_path.exists(), "Data file does not exist."
```

```python
stanford_data_path = Path(
    "../../data/radgraph-xl-a-large-scale-expert-annotated-dataset-for-entity-and-relation-extraction-from-radiology-reports-1.0.0/stanford-radgraph-XL.jsonl"
)
assert stanford_data_path.exists(), "Data file does not exist."
```

```python
mimic_data= []
with open(mimic_data_path, 'r') as f:
    for line in f:
        item = json.loads(line)
        mimic_data.append(item)

stanford_data = []
with open(stanford_data_path, 'r') as f:
    for line in f:
        item = json.loads(line)
        stanford_data.append(item)
```

```python
print(len(mimic_data))
print(len(stanford_data))
```
